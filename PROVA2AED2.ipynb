```python
# função com falhas



def selection_sort(arr):
    for i in range(len(arr)-1): #percorre de 0 a 7
        min_idx = i # MENOR VALOR DA LISTA (estava errado pois, i + 1 já pega automaticamente o valor proximo a i, mas pode ser i ja seja o menor)
        for j in range(i+1, len(arr)): # j faz iteração (não faz sentido j iterar com i, pois já foi colocado na variavel min_idx)
            if arr[j] < arr[min_idx]: #menor ou igual (estava errado pois arr[j] nao compara com arr[i], mas sim o ultimo menor da vez)
                min_idx = j # variavel min_idx recebe o menor valor

        arr[i], arr[min_idx] = arr[min_idx], arr[i] # aqui a troca era impossivel, entao fiz a troca ao mesmo tempo para nao perder valores
                                                    # tambem era possivel salvar em uma variavel
        
    return arr
```
